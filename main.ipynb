```python
import pandas as pd
```

```python
df = pd.read_json('csm.json')
df.head()
```

```python
df.info()
```

```python
# extract around 5000 from the dataset
sample = df.iloc[:5000]
sample
```

```python
selected = sample[['title','creation_date','url','images','creators']]
selected
```

```python
# define an artifact
class Artifact:    
    def __init__(self, data):
        print(data)

        self.title = data['title']
        self.date = data['creation_date']
        self.url= data['url']
        self.creators = [c['description'] for c in data['creators'] if 'description' in c] #extract only the key 'description' within the creators attribute

    def __str__(self): # string representation of the class / object
        return f"title: {self.title}\n creators:{self.creators}"
```

```python
# a is test object of the class artifact
a = Artifact(selected.iloc[0]) # print the first row 
print(a)
```

```python
from rdflib import Graph
from rdflib.namespace import RDF, RDFS
from rdflib import Namespace
```

```python
# define and bind namespace
CRM = Namespace("http://www.cidoc-crm.org/cidoc-crm/")
DC = Namespace("http://purl.org/dc/elements/1.1/")
EX = Namespace("http://w3id.org/example/")

g = Graph()

g.bind("crm", CRM)
g.bind("dc", DC)
g.bind("ex", EX)

```
